```python
import numpy as np
import pandas as pd
import os
import shutil

fpath = "/home/arttu/Documents/research/melanoma/"
```

### Originally, the image data is in one train folder and the labels are found from a separate csv file. We need to divide the images from the single folder to different folders based on their labels which are listed in the csv file.

```python
labels = pd.read_csv(fpath+'labels.csv').set_index('image')
labels
```

```python
imgs = os.listdir(fpath+'data/images')
len(imgs)
```

```python
src_path = "/home/arttu/Documents/research/melanoma/data/images/"
fpath = "/home/arttu/Documents/research/melanoma/data/"
for img in imgs:
    if labels.loc[img[:-4], 'MEL'] == 1.0:
        dst_path = fpath+'MEL/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'NV'] == 1.0:
        dst_path = fpath+'NV/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'BCC'] == 1.0:
        dst_path = fpath+'BCC/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'AK'] == 1.0:
        dst_path = fpath+'AK/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'BKL'] == 1.0:
        dst_path = fpath+'BKL/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'DF'] == 1.0:
        dst_path = fpath+'DF/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'VASC'] == 1.0:
        dst_path = fpath+'VASC/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'SCC'] == 1.0:
        dst_path = fpath+'SCC/'
        shutil.move(src_path+img, dst_path+img)
    elif labels.loc[img[:-4], 'UNK'] == 1.0:
        dst_path = fpath+'UNK/'
        shutil.move(src_path+img, dst_path+img)
```
